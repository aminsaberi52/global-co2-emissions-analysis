# Global CO₂ Emissions Analysis (1960–2022)

This notebook presents a structured analysis of global CO₂ emissions using the 
OWID CO₂ dataset. The study focuses on long-term emission trends, per-capita 
patterns, economic drivers, and cross-country differences, covering the period 
1960–2022. All visualizations and analytical steps are fully reproducible.

```python
# Import required Python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Data Preparation

In this section, the original OWID CO₂ dataset is loaded, filtered, and cleaned.
Only variables relevant to the analytical framework are retained, and the dataset 
is restricted to the selected countries and the study period (1960–2022).

```python
# 1. Load the original OWID CO₂ dataset
df_raw = pd.read_csv("../../co2-data/owid-co2-data.csv")

# 2. Select only the columns needed for the analysis
cols = [
    "country",
    "year",
    "co2",             # total CO₂ emissions (million tonnes)
    "co2_per_capita",  # CO₂ emissions per person (tonnes)
    "population",
    "gdp"              # Gross Domestic Product (USD, current)
]
df = df_raw[cols].copy()

# 3. Select the countries/entities used in the study
selected_entities = [
    "World",
    "China",
    "United States",
    "India",
    "Russia",
    "Japan",
    "Iran"
]

df = df[df["country"].isin(selected_entities)].copy()

# 4. Limit the time range to 1960–2022
df = df[(df["year"] >= 1960) & (df["year"] <= 2022)].copy()

# 5. Remove rows where CO₂ or per-capita values are missing
df = df.dropna(subset=["co2", "co2_per_capita"])

# 6. Save the cleaned dataset for future use
df.to_csv("clean_co2_dataset.csv", index=False)

df.head()
```

```python
# Helper function to save and display figures
def save_and_show(fig, filename, dpi=300):
    """
    fig: Matplotlib figure object
    filename: output file name (PNG)
    dpi: resolution of image
    """
    fig.tight_layout()
    fig.savefig(filename, dpi=dpi, bbox_inches="tight")
    plt.show()
```

## 2. CO₂ Emission Trends

The following plots illustrate the long-term evolution of CO₂ emissions across key 
countries. Total emissions and per-capita emissions are examined separately to 
highlight differences between high-population and high-consumption economies.

```python
# Line chart for total CO₂ emissions (1960–2022)
fig, ax = plt.subplots(figsize=(12, 7))

for country in selected_entities:
    sub = df[df["country"] == country]
    ax.plot(sub["year"], sub["co2"], label=country)

ax.set_title("CO₂ Emissions Over Time (1960–2022)", fontsize=18)
ax.set_xlabel("Year", fontsize=15, fontweight="bold")
ax.set_ylabel("CO₂ emissions (million tonnes)", fontsize=15)

# 🔥 Bigger, clearer numbers
ax.set_xticks(range(1960, 2023, 10))
ax.tick_params(axis="x", labelsize=14)
ax.tick_params(axis="y", labelsize=14)

ax.grid(True, alpha=0.3)
ax.legend(fontsize=12)

save_and_show(fig, "fig1_trend_total_co2.png")
```

```python
# Line chart for CO₂ per capita (1960–2022)
fig, ax = plt.subplots(figsize=(12, 7))

# Plot countries
for country in selected_entities:
    sub = df[df["country"] == country]
    ax.plot(sub["year"], sub["co2_per_capita"], label=country)

# Compute average inside the same cell
avg_df = (
    df[df["country"] != "World"]
    .groupby("year")["co2_per_capita"]
    .mean()
    .reset_index()
)

# Plot average line
ax.plot(
    avg_df["year"], avg_df["co2_per_capita"],
    label="Average", linewidth=3.5, color="black", linestyle="--"
)

ax.set_title("CO₂ Emissions Per Capita (1960–2022)", fontsize=18)
ax.set_xlabel("Year", fontsize=15, fontweight="bold")
ax.set_ylabel("CO₂ per capita (tonnes per person)", fontsize=15)

# 🔥 Bigger, clearer numbers
ax.set_xticks(range(1960, 2023, 10))
ax.tick_params(axis="x", labelsize=14)
ax.tick_params(axis="y", labelsize=14)

ax.grid(True, alpha=0.3)
ax.legend(fontsize=12)

save_and_show(fig, "fig2_trend_co2_per_capita.png")
```

## 3. Country Comparisons in the Latest Year

These visualizations compare total and per-capita CO₂ emissions in the most recent 
year available in the dataset. The global aggregate (World) is relabeled as 
"Average" to improve interpretability.

```python
# Prepare latest-year data (local to this cell)
latest_year = df["year"].max()
df_latest = df[(df["year"] == latest_year) & (df["country"] != "World")] \
    .sort_values("co2", ascending=False)

# Bar chart (total CO₂) — readability optimized
fig, ax = plt.subplots(figsize=(9, 5))

ax.barh(df_latest["country"], df_latest["co2"])
ax.set_title(f"CO₂ Emissions by Country in {latest_year}", fontsize=14)
ax.set_xlabel("CO₂ emissions (million tonnes)", fontsize=12)
ax.set_ylabel("Country", fontsize=12, fontweight="bold")

ax.tick_params(axis="x", labelsize=11)
ax.tick_params(axis="y", labelsize=11)
ax.invert_yaxis()
ax.grid(axis="x", alpha=0.3)

save_and_show(fig, "fig3_bar_latest_year_total_co2.png")
```

```python
# Prepare latest-year per-capita data (local to this cell)
latest_year = df["year"].max()
df_latest_pc = df[df["year"] == latest_year] \
    .sort_values("co2_per_capita", ascending=False)

labels = df_latest_pc["country"].tolist()
values = df_latest_pc["co2_per_capita"].tolist()
colors = ["#1f77b4"] * len(labels)

# Replace "World" label with "Average"
if "World" in labels:
    world_idx = labels.index("World")
    labels[world_idx] = "Average"
    colors[world_idx] = "#2EC4B6"

# Bar chart — readability optimized
fig, ax = plt.subplots(figsize=(9, 5))

ax.barh(labels, values, color=colors)
ax.set_title(f"CO₂ Per Capita by Country in {latest_year}", fontsize=14)
ax.set_xlabel("CO₂ per capita (tonnes per person)", fontsize=12)
ax.set_ylabel("Country", fontsize=12, fontweight="bold")

ax.tick_params(axis="x", labelsize=11)
ax.tick_params(axis="y", labelsize=11)
ax.invert_yaxis()
ax.grid(axis="x", alpha=0.3)

save_and_show(fig, "fig3b_bar_latest_year_per_capita.png")
```

## 4. Relationship Between GDP and CO₂ Emissions

This section examines the economic determinants of CO₂ emissions through a 
scatter plot that relates national GDP to CO₂ emissions per capita.

```python
# Relationship between GDP and CO₂ per capita (latest year)
df_latest_gdp = df[(df["year"] == latest_year) & df["gdp"].notna()].copy()

fig, ax = plt.subplots(figsize=(10, 6))

# Scatter points
ax.scatter(df_latest_gdp["gdp"], df_latest_gdp["co2_per_capita"], s=100)

# Add labels for each point
for _, row in df_latest_gdp.iterrows():
    ax.text(row["gdp"], row["co2_per_capita"], row["country"], fontsize=9)

ax.set_title(f"GDP vs CO₂ Per Capita ({latest_year})")
ax.set_xlabel("GDP (current US$)")
ax.set_ylabel("CO₂ per capita (tonnes per person)")
ax.grid(True, alpha=0.3)

save_and_show(fig, "fig4_scatter_gdp_vs_co2_per_capita.png")
```

## 5. Correlation Analysis

To better understand the interplay among demographic, economic, and emission-related 
variables, a correlation matrix is computed and visualized as a heatmap. This provides 
a high-level overview of the structural relationships in the data.

```python
# Correlation heatmap using values from the latest year
df_corr = df[df["year"] == latest_year][["co2", "co2_per_capita", "population", "gdp"]].dropna()

corr = df_corr.corr()

fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", ax=ax)

ax.set_title(f"Correlation Matrix ({latest_year})")

save_and_show(fig, "fig5_correlation_heatmap.png")
```

## 6. Conclusion

The analysis reveals persistent increases in global CO₂ emissions, substantial 
cross-country inequalities, and strong links between economic development and 
emission intensity. These findings underscore the complexity of global climate
